```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(11):
    j = 10
    while j <= 10000:
        dots = np.loadtxt(f"./output_squares/dots/dots_squares{i}_{j}.txt").T
        width = abs(dots[0][0] - dots[0][1]) * 0.9

        a_0 = np.loadtxt(
            f"./output_squares/output_squares{i}.txt", skiprows=13, max_rows=1)
        a_n = np.loadtxt(
            f"./output_squares/output_squares{i}.txt", skiprows=14, max_rows=1)
        b_n = np.loadtxt(
            f"./output_squares/output_squares{i}.txt", skiprows=15, max_rows=1)

        l = np.pi
        y = a_0 / (2 * l) + np.array([np.sum([coef1 * np.cos(ses * (k1 + 1) * (np.pi / l))
                                              for k1, coef1 in enumerate(a_n / l)]) + np.sum([coef2 * np.sin(ses * (k + 1) * (np.pi / l))
                                                                                              for k, coef2 in enumerate(b_n / l)])
                                      for ses in dots[0]])

        fig, ax = plt.subplots()
        ax.plot(dots[0], y)

        colors = np.random.choice(100, size=(dots[0].size, 3)) / 100
        ax.bar(dots[0], dots[1], align='edge', width=width, color=colors)
        plt.show()

        j *= 10
```

```python
for i in range(11):
    j = 10
    while j <= 10000:
        dots = np.loadtxt(f"./output_trapezoids/dots/dots_trapezoids{i}_{j}.txt").T
        width = abs(dots[0][0] - dots[0][1]) * 0.9


        a_0 = np.loadtxt(
            f"./output_trapezoids/output_trapezoids{i}.txt", skiprows=13, max_rows=1)
        a_n = np.loadtxt(
            f"./output_trapezoids/output_trapezoids{i}.txt", skiprows=14, max_rows=1)
        b_n = np.loadtxt(
            f"./output_trapezoids/output_trapezoids{i}.txt", skiprows=15, max_rows=1)

        l = np.pi
        y = a_0 / (2 * l) + np.array([np.sum([coef1 * np.cos(ses * (k1 + 1) * (np.pi / l))
                                              for k1, coef1 in enumerate(a_n / l)]) + np.sum([coef2 * np.sin(ses * (k + 1) * (np.pi / l))
                                                                                              for k, coef2 in enumerate(b_n / l)])
                                      for ses in dots[0]])

        fig, ax = plt.subplots()

        colors = np.random.choice(100, size=(dots[0].size, 3)) / 100

        for k in range(dots[0].size):
            ax.fill_between(dots[0][k:k+2], dots[1][k:k+2], color=colors[k])
        ax.plot(dots[0], y)
        plt.show()

        j *= 10
```
